Basic Operations on Images

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('MY.jpg')
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
k = cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
px = img[100,100]
print(px)
```

```python
img[100,100] = [255, 255, 255]
```

```python
img[100:200, 100:200] = [0, 255, 0]
```

```python
img[100:200, 100:200, 2] = 255
```

```python
img[200:400, 200:400, 2] = 255
```

```python
blue = img[100,100,0]
print(blue)
```

```python
img[100,100] = [255,255,255]
print(img[100,100])
```

```python
img.item(10,10,2)
```

```python
img.itemset((10,10,2),100)
img.item(10,10,2)
```
